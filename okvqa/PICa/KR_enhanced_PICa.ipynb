```python
import requests, json, time
from tqdm import tqdm
import os
os.environ['https_proxy'] = 'http://127.0.0.1:15777'
os.environ['http_proxy'] = 'http://127.0.0.1:15777'

# obj.keys()
# obj['edges'][0]['surfaceText']
with open("./projects/data122/vscode_home/projects/PICa/tag2description_conceptnet.json", 'r', encoding='utf-8') as f:
    tag2description_new = json.load(f)

with open("./projects/data122/vscode_home/projects/PICa/tag2description.json", 'r', encoding='utf-8') as f:
    tag2description = json.load(f)
for tag in tqdm(tag2description, total=len(tag2description)):
    if tag2description_new[tag] == "":
        tag2description_new[tag] = tag
    if tag in tag2description_new:
        continue
    try:
        obj = requests.get(f'http://api.conceptnet.io/c/en/{tag}').json()
        if 'error' in obj.keys():
            desc = tag
        else:
            desc = [obj['edges'][i]['surfaceText'] for i in range(len(obj['edges'])) if obj['edges'][i]['surfaceText'] is not None]
            desc ='\n'.join(desc)
        print(tag, desc)
        tag2description_new[tag] = desc
    except Exception as e:
        print(e)
        time.sleep(5)
        continue

with open("./projects/data122/vscode_home/projects/PICa/tag2description_conceptnet.json", 'w', encoding='utf-8') as f:
    json.dump(tag2description_new, f, ensure_ascii=False, indent=4)
```
